```python
def check(Dimension, test_size, x, y):
    r2 = 0.889063290910722
    for column in x:
        col_name = f"{column}_{Dimension}"
        new_col = x[column] ** Dimension
        x[col_name] = new_col
        model = LinearRegression()
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=0)
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        
        r2_new = metrics.r2_score(y_test, y_pred)
        
        if r2_new < r2:
            x.drop(columns=[col_name], inplace=True)
        else:
            r2 = r2_new
    
    print(r2)

# Example usage:
# Assuming x and y are your feature and target DataFrames
# Dimension is the number of features, and test_size is the proportion of the dataset to include in the test split
check(Dimension=2, test_size=0.2, x=x, y=y)        
        
```
